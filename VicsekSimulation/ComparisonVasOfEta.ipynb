## Production of the plot comparing the calculated data to the data from the paper 

Data from the paper was extracted via the online tool PlotDigitizer. 

```python
import numpy as np
import VicsekSimulationNumba as vi
import matplotlib.pyplot as plt 
import matplotlib
import pandas as pd
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator, AutoLocator)
```

```python
dataPaper = pd.read_csv("Data/vaOfEta40paper.csv")
dataCalc = pd.read_csv("Data/vaOfEta40calculatedRho40WithErrors.csv")
dataPaper2 = pd.read_csv("Data/vaOfEta100paper.csv")
dataCalc2 = pd.read_csv("Data/vaOfEta100calculatedRho40WithErrors.csv")
dataPaper3 = pd.read_csv("Data/vaOfEta400paper.csv")
dataCalc3 = pd.read_csv("Data/vaOfEta400calculatedRho40WithErrors.csv")
```

```python
xpaper = np.array(dataPaper)[:,0]
ypaper = np.array(dataPaper)[:,1]
xCalc = np.array(dataCalc)[:,2]
yCalc = np.array(dataCalc)[:,1]
error = np.array(dataCalc)[:,3]
xpaper2 = np.array(dataPaper2)[:,0]
ypaper2 = np.array(dataPaper2)[:,1]
xCalc2 = np.array(dataCalc2)[:,2]
yCalc2 = np.array(dataCalc2)[:,1]
error2 = np.array(dataCalc2)[:,3]
xpaper3 = np.array(dataPaper3)[:,0]
ypaper3 = np.array(dataPaper3)[:,1]
xCalc3 = np.array(dataCalc3)[:,2]
yCalc3 = np.array(dataCalc3)[:,1]
error3 = np.array(dataCalc3)[:,3]
```

```python
matplotlib.use("pgf")
matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})
```

```python
fig = plt.figure()
fig.set_size_inches(w=5.78851, h=3.5)
ax1 = fig.add_subplot(1,1,1)

ax1.set_xlim([0,5])
ax1.set_ylim([0,1])

#ax1.grid( which="both")

#ax1.xaxis.set_minor_locator(AutoMinorLocator())
#ax1.xaxis.set_major_locator(AutoLocator())
#ax1.yaxis.set_minor_locator(AutoMinorLocator())
#ax1.yaxis.set_major_locator(AutoLocator())

#ax1.tick_params(which='both', width=1)
#ax1.tick_params(which='major', length=7)
#ax1.tick_params(which='minor', length=4, grid_linestyle="--", grid_color="lightgray")

ax1.set_xlabel('$\eta$')
ax1.set_ylabel('$v_a$')

plt.scatter(xpaper,ypaper, c = 'grey', label = 'N = 40', marker = 's', s = 35)
plt.errorbar(xCalc, yCalc, yerr = error, label = 'N = 40', fmt = 'o', ecolor = 'red')
plt.scatter(xpaper2,ypaper2, c = 'grey', marker = '+', label = 'N = 100',s = 60)
plt.errorbar(xCalc2, yCalc2, yerr = error2, label = 'N = 100', fmt = 'o', ecolor = 'red')
plt.scatter(xpaper3,ypaper3, c = 'grey', marker = 'x', label = 'N = 400',s = 40)
plt.errorbar(xCalc3, yCalc3,  yerr = error3, label = "N = 400", fmt = 'o', ecolor = 'red')

plt.legend()

plt.savefig('vaOfEtaCalc.pgf')
```
