# Reproduction of the first Figure in the Vicsek paper

Produces one plot with the four subplots.

```python
import VicsekSimulationNumba as vi
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
v = vi.Vicsek(300, 7, 2, True)
v2 = vi.Vicsek(300, 25, 0.1, False)
v3 = vi.Vicsek(300, 5, 0.1, False)

```

```python
vD = (v.x_i,v.v_i,v.theta_i)
```

```python
for j in range(500):
                x_ip1, theta_ip1, v_ip1 = vi.numbaUpdate(v.N, v.varT, v.L, v.x_i, v.v_i, v.theta_i, v.eta)
                v.x_i, v.theta_i, v.v_i = x_ip1, theta_ip1, v_ip1
                v.theta_i = np.where(v.theta_i > 0, v.theta_i, v.theta_i + 2*np.pi)
```

```python
for j in range(500):
                x_ip1, theta_ip1, v_ip1 = vi.numbaUpdate(v2.N, v2.varT, v2.L, v2.x_i, v2.v_i, v2.theta_i, v2.eta)
                v2.x_i, v2.theta_i, v2.v_i = x_ip1, theta_ip1, v_ip1
                v2.theta_i = np.where(v2.theta_i > 0, v2.theta_i, v2.theta_i + 2*np.pi)
```

```python
for j in range(500):
                x_ip1, theta_ip1, v_ip1 = vi.numbaUpdate(v3.N, v3.varT, v3.L, v3.x_i, v3.v_i, v3.theta_i, v3.eta)
                v3.x_i, v3.theta_i, v3.v_i = x_ip1, theta_ip1, v_ip1
                v3.theta_i = np.where(v3.theta_i > 0, v3.theta_i, v3.theta_i + 2*np.pi)
```

```python
plt.quiver(v.x_i[:,[0]],v.x_i[:,[1]],v.v_i[:,[0]],v.v_i[:,[1]], color = cm.get_cmap('viridis')(v.theta_i/(2*np.pi)))
```

```python
plt.quiver(v2.x_i[:,[0]],v2.x_i[:,[1]],v2.v_i[:,[0]],v2.v_i[:,[1]], color = cm.get_cmap('viridis')(v2.theta_i/(2*np.pi)))
```

```python
plt.quiver(v3.x_i[:,[0]],v3.x_i[:,[1]],v3.v_i[:,[0]],v3.v_i[:,[1]], color = cm.get_cmap('viridis')(v3.theta_i/(2*np.pi)))
```

```python
fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(10)
ax = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax.set_title('(a)', loc="left")
ax2.set_title('(b)', loc="left")
ax3.set_title('(c)', loc="left")
ax4.set_title('(d)', loc="left")
ax.quiver(vD[0][:,[0]],vD[0][:,[1]],vD[1][:,[0]],vD[1][:,[1]], color = cm.get_cmap('viridis')(vD[2]/(2*np.pi)))
ax2.quiver(v2.x_i[:,[0]],v2.x_i[:,[1]],v2.v_i[:,[0]],v2.v_i[:,[1]], color = cm.get_cmap('viridis')(v2.theta_i/(2*np.pi)))
ax3.quiver(v.x_i[:,[0]],v.x_i[:,[1]],v.v_i[:,[0]],v.v_i[:,[1]], color = cm.get_cmap('viridis')(v.theta_i/(2*np.pi)))
ax4.quiver(v3.x_i[:,[0]],v3.x_i[:,[1]],v3.v_i[:,[0]],v3.v_i[:,[1]], color = cm.get_cmap('viridis')(v3.theta_i/(2*np.pi)))
```
